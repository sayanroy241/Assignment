1.  What is K-Nearest Neighbors (KNN) and how does it work?

K-Nearest Neighbors (KNN) is a non-parametric, supervised machine learning algorithm used for both classification and regression tasks. It operates on the principle that similar data points exist close to each other in the feature space—essentially, "birds of a feather flock together".

KNN is considered a "lazy learning" algorithm because it does not build a model during a training phase; instead, it memorizes the entire training dataset and performs all computations during prediction. This makes it simple to implement but potentially computationally intensive for large datasets.

How Does KNN Work?
Step-by-step process:

Choose the value of K:
Decide how many neighbors (K) you want the algorithm to consider when making a prediction.

Calculate distances:
For a new (unlabeled) data point, compute the distance between this point and every point in the training dataset. Common distance metrics include Euclidean, Manhattan, Minkowski, and others.

Find the K nearest neighbors:
Identify the K data points in the training set that are closest to the new point based on the chosen distance metric.

Make a prediction:

For classification: Assign the class label that is most common (the mode) among the K nearest neighbors (majority or plurality voting).

For regression: Compute the average (or sometimes the weighted average) of the values of the K nearest neighbors.

Assign the result:
The predicted class or value is assigned to the new data point.

4.  What is the Curse of Dimensionality in KNN?

Curse of Dimensionality in KNN
The "curse of dimensionality" refers to the set of problems and phenomena that arise when analyzing and organizing data in high-dimensional spaces. In the context of K-Nearest Neighbors (KNN), the curse of dimensionality has significant negative effects:

Distance Becomes Less Meaningful:
As the number of features (dimensions) increases, the distance between any two data points becomes increasingly similar. This means that all points tend to become almost equidistant from each other, making it difficult for the KNN algorithm to distinguish between close and distant neighbors effectively.

Sparsity of Data:
High-dimensional spaces are mostly empty, so data points become sparse. This sparsity reduces the likelihood that any two points are truly "near" each other, undermining the core assumption of KNN that similar points are close in feature space.

Increased Computational Cost:
The number of calculations required grows rapidly with the number of dimensions, making KNN computationally expensive and less practical for high-dimensional datasets.

Overfitting Risk:
KNN is more prone to overfitting in high dimensions because the algorithm may find spurious patterns in the noise of the data rather than genuine similarities.

5. How can we choose the best value of K in KNN?

How to Choose the Best Value of K in KNN
Selecting the optimal value of K (the number of neighbors) is crucial for the performance of the K-Nearest Neighbors algorithm. Here are widely used strategies:

Cross-Validation:
The most common and effective approach is to use cross-validation. This involves:

Trying a range of K values (e.g., 1 to 30).

For each K, evaluating the model’s performance (such as accuracy for classification or mean squared error for regression) using cross-validation splits.

Selecting the K that yields the best average performance across the splits.

Odd Values for Classification:
For binary classification tasks, it’s recommended to use an odd value for K to avoid ties when voting for the predicted class.

Heuristic and Data-Dependent Approaches:

Larger K values can reduce the effect of noise but may make class boundaries less distinct.

Smaller K values can capture local structure but are more sensitive to noise and outliers.

The optimal K is highly data-dependent and may require experimentation.

Visualization:
Plotting the model’s performance against different K values can help visualize where the accuracy or error stabilizes, making it easier to select an optimal K.

Advanced Variants:
Some adaptive KNN algorithms dynamically select K for each data point based on local data characteristics, but these are more complex and less commonly used in standard practice.

6.  What are KD Tree and Ball Tree in KNN?

KD Tree and Ball Tree in KNN
KD Tree and Ball Tree are advanced data structures used to accelerate nearest neighbor searches in KNN, especially when dealing with large datasets. Both structures organize data to reduce the number of distance calculations required, making KNN much more efficient.

KD Tree (K-Dimensional Tree)
Structure:
A KD Tree is a binary tree that recursively partitions the data space along different feature axes. At each level, the data is split by the median value of one dimension, cycling through all available dimensions.

Construction:
The tree is built by choosing a splitting axis (which cycles through the dimensions) and dividing the data at the median value along that axis. This process continues recursively until each region contains a small number of points or a single point.

Querying:
To find the nearest neighbors, the tree is traversed from the root to a leaf node, following the path that best matches the query point. After reaching a leaf, the algorithm may backtrack to check other branches if they could contain closer points, using a branch-and-bound approach.

Performance:
KD Trees are highly effective for low- to moderate-dimensional data but become less efficient as dimensionality increases due to the curse of dimensionality.

Use in KNN:
KD Trees significantly speed up neighbor searches compared to brute-force methods, especially for large datasets with a moderate number of features.

Ball Tree
Structure:
A Ball Tree organizes data points in a series of nested hyperspheres ("balls") instead of axis-aligned splits. Each node in the tree represents a ball that contains a subset of the data points.

Construction:
The data is recursively split into two groups, each enclosed in its own ball, with the process continuing until each ball contains a small number of points.

Querying:
When searching for nearest neighbors, the algorithm quickly eliminates entire balls if it can determine that none of their points could be closer than the current best candidates, thus reducing unnecessary distance calculations.

Performance:
Ball Trees are often preferred for higher-dimensional data or when the data is not well-suited for axis-aligned splits, as they can handle complex data distributions more flexibly than KD Trees.


7.  When should you use KD Tree vs. Ball Tree?

When to Use:
Use KD Tree When:
You have low-dimensional data (usually < 20 dimensions).

Your features are numerical.

You're using axis-aligned distance metrics (like Euclidean).

You want fast queries on small to medium datasets.

Use Ball Tree When:
You have medium- to high-dimensional data (e.g., 20–100+ dimensions).

You need to support non-Euclidean distance metrics.

Your data is not well-separated along axis-aligned cuts.

You’re dealing with sparse or irregularly shaped clusters.

8.  What are the disadvantages of KNN?

Disadvantages of KNN
Computationally Intensive:
KNN is a lazy learning algorithm, meaning it does not build a model during training but instead computes distances to all training points at prediction time. This makes it slow and resource-intensive, especially for large datasets.

High Memory Consumption:
The entire training dataset must be stored in memory for predictions, leading to high memory usage as data size grows.

Scalability Issues:
As the number of training samples and features increases, both computation time and memory requirements grow rapidly, making KNN impractical for big data unless optimized structures (like KD Tree or Ball Tree) or approximate methods are used.

Curse of Dimensionality:
KNN performs poorly with high-dimensional data because distances between points become less meaningful, and all points tend to appear similarly distant, reducing the algorithm’s effectiveness.

Sensitive to Irrelevant Features and Feature Scale:
All features are considered equally in distance calculations. Irrelevant features or features with larger scales can dominate results, leading to inaccurate predictions unless features are properly selected and scaled.

No Model Interpretability or Feature Importance:
KNN does not produce a model or provide insights into feature importance, making it less interpretable than other algorithms.

Difficult Choice of K:
The algorithm’s performance is sensitive to the choice of K (number of neighbors). An inappropriate K can lead to overfitting or underfitting.

Struggles with Imbalanced Classes:
KNN can perform poorly when classes are imbalanced, as the majority class can dominate predictions in local neighborhoods.

9.  How does feature scaling affect KNN?

How Feature Scaling Affects KNN
Feature scaling is crucial for the K-Nearest Neighbors (KNN) algorithm because KNN relies on distance calculations to determine the similarity between data points. Here’s how feature scaling impacts KNN:

Prevents Dominance by Large-Scale Features:
If features have different scales (e.g., income in thousands vs. age in years), features with larger numerical ranges will dominate the distance computation. This can lead to biased or incorrect predictions, as KNN will prioritize differences in those features over others, regardless of their true importance.

Ensures Fair Contribution of All Features:
Scaling transforms all features to a comparable range, ensuring each feature contributes equally to the distance metric. This makes the KNN algorithm more robust and accurate, as no single feature can disproportionately influence the outcome.

Improves Model Performance:
Studies and practical experiments show that feature scaling significantly improves KNN’s accuracy and reliability. Different scaling methods (such as standardization or min-max scaling) can be tested, but standardization (mean = 0, standard deviation = 1) is commonly effective.

Consistency Between Training and Test Data:
It is important to fit the scaler on the training data and use the same transformation on the test data to maintain consistency and prevent data leakage.

10.  What is PCA (Principal Component Analysis)?

Principal Component Analysis (PCA) is a statistical and machine learning technique used for dimensionality reduction—that is, simplifying large, complex datasets by transforming them into a smaller set of variables (called principal components) that retain most of the original information.

11.  How does PCA work?

How PCA Works
Transforms Data:
PCA transforms potentially correlated variables into a new set of uncorrelated variables (principal components), ordered by the amount of variance (information) they capture from the original dataset.

Captures Variance:
The first principal component captures the largest possible variance in the data. Each subsequent component captures the next highest variance, with all components being orthogonal (uncorrelated) to each other.

Reduces Complexity:
By keeping only the top principal components, PCA reduces the number of features, making data easier to visualize, analyze, and process for machine learning, while minimizing information loss.

12. What is the geometric intuition behind PCA?

Principal Component Analysis (PCA) is a powerful dimensionality reduction technique that finds a new set of dimensions (called principal components) for your data. The geometric intuition behind PCA can be understood by thinking of your data as a cloud of points in a multi-dimensional space.

Here's the breakdown of the geometric intuition:

1.Mean-Centering: Before anything else, PCA typically starts by mean-centering the data. This means shifting the entire cloud of points so that its center (the mean of all data points) aligns with the origin of the coordinate system. This is done to remove any arbitrary bias in the measurements and focus purely on the spread and relationships within the data.

2.Finding the Direction of Maximum Variance (First Principal Component):

Imagine you have a scatter plot of data points in 2D (e.g., height vs. weight). These points form an elliptical shape.
PCA's first goal is to find the direction (a line) through this cloud of points along which the data varies the most.
Think of it like trying to fit a line to the data such that if you project all the data points onto this line, the spread (variance) of these projected points is maximized.
This "best-fit" line is the first principal component (PC1). It captures the most significant pattern of variation in your data. Geometrically, it's the major axis of the ellipsoid that best approximates the data cloud.

3.Finding the Next Direction of Maximum Variance (Second Principal Component):

Once PC1 is determined, PCA looks for the next direction that captures the most remaining variance.
Crucially, this second direction (the second principal component, PC2) must be orthogonal (perpendicular) to the first principal component.
Again, you're looking for a line such that projecting the data onto it maximizes the variance of the projected points, given the constraint that it's perpendicular to PC1.
In a 2D example, if PC1 is the major axis of the ellipse, PC2 would be the minor axis.

4.Extending to Higher Dimensions:

This process continues for subsequent principal components. Each new principal component is chosen to be orthogonal to all previously found principal components and to capture the maximum remaining variance in the data.

If you have 'p' original dimensions, you can find up to 'p' principal components.

5.Projection and Dimensionality Reduction:

Once the principal components are found, you can project your original data onto a subset of these new dimensions.
Since the principal components are ordered by the amount of variance they explain, you can choose to keep only the first few principal components that capture most of the important information.
This projection effectively transforms your data from its original high-dimensional space into a lower-dimensional space, while preserving as much of the original data's variance (and thus information) as possible.

14.  What are Eigenvalues and Eigenvectors in PCA?

Eigenvectors and eigenvalues are fundamental to how Principal Component Analysis (PCA) works:

Eigenvectors:
In PCA, the eigenvectors of the data’s covariance matrix define the directions (axes) of the new feature space—these are the principal components. Each eigenvector points in a direction where the data varies the most, and all principal components are orthogonal (uncorrelated) to each other.

Eigenvalues:
Each eigenvector has a corresponding eigenvalue, which indicates how much variance (information) from the original data is captured along that principal component. The larger the eigenvalue, the more variance is explained by its associated eigenvector.

15. How do you decide the number of components to keep in PCA?

Selecting the number of principal components to retain in PCA is a balance between reducing dimensionality and preserving as much information (variance) as possible. Several well-established methods are commonly used:

1. Cumulative Explained Variance Threshold
Approach: Choose the smallest number of components such that their cumulative explained variance exceeds a predefined threshold, often 80%, 90%, or 95% of the total variance.

Rationale: This ensures that the retained components capture most of the original data's information, while discarding components that mainly represent noise or minor variations.

Example: If you set a 95% threshold, keep enough components so that together they explain at least 95% of the variance in the data.

2. Scree Plot (Elbow Method)
Approach: Plot the eigenvalues (or explained variance) of each principal component in descending order. Look for an "elbow" point where the explained variance drops off sharply; keep components up to this point.

Rationale: The elbow indicates diminishing returns—additional components contribute little extra information.

3. Kaiser’s Rule
Approach: Retain all components with eigenvalues greater than or equal to 1.

Rationale: An eigenvalue of 1 means the component explains as much variance as an original variable; components with lower eigenvalues are considered less informative.

4. Parallel Analysis and Other Statistical Tests
Approach: Compare the eigenvalues from your data to those from randomly generated data; retain components with eigenvalues higher than those expected by chance.

Rationale: This method is more robust and less subjective than the scree plot or Kaiser’s rule, especially for complex datasets.

5. Interpretability
Approach: Beyond statistical criteria, consider whether the retained components have meaningful or interpretable loading patterns.

Rationale: Sometimes, components that are statistically significant may not be useful if they lack a clear interpretation in the context of your analysis.

16. Can PCA be used for classification?


Yes, PCA (Principal Component Analysis) can be used in classification tasks, but with a specific role: PCA itself is not a classification algorithm; it is a dimensionality reduction technique that is often used as a preprocessing step before classification.

How PCA Supports Classification
Feature Extraction/Reduction:
PCA transforms the original high-dimensional data into a smaller set of uncorrelated variables (principal components) that retain most of the variance (information) from the original data. This helps reduce noise, redundancy, and computational complexity, making the data more manageable for classification algorithms.

Improved Classifier Performance:
By reducing the number of features, PCA can help classifiers (such as k-nearest neighbors, decision trees, or neural networks) perform better, especially when the original dataset has many correlated or irrelevant features. This is particularly effective for algorithms that do not have built-in feature selection mechanisms.

Visualization:
PCA can project high-dimensional data into two or three principal components, making it easier to visualize class separations and patterns before applying a classifier.

17.  What are the limitations of PCA?

Limitations of PCA (Principal Component Analysis)
PCA is a powerful tool for dimensionality reduction, but it comes with several important limitations:

Loss of Interpretability:
Principal components are linear combinations of original features, making them difficult to interpret. After transformation, it can be challenging to relate principal components back to the original variables, which complicates explaining results to stakeholders.

Information Loss:
Reducing dimensionality inevitably leads to some loss of information, especially if too few principal components are retained. The trade-off between simplification and retaining important data characteristics must be carefully managed.

Assumption of Linearity:
PCA assumes that the relationships between variables are linear. If the actual data structure is nonlinear, PCA may not capture important patterns, leading to suboptimal results.

Sensitivity to Feature Scaling:
PCA is sensitive to the scale of the data. Features with larger variances can dominate the principal components. Therefore, data normalization or standardization is essential before applying PCA; otherwise, the results may be misleading.

Computational Complexity:
The computational cost of PCA is cubic with respect to the number of features, making it expensive and sometimes impractical for very high-dimensional datasets.

Sensitivity to Outliers and Missing Data:
PCA can be heavily influenced by outliers and struggles with missing data, which can distort the principal components and the variance they capture.

Orthogonality Constraint:
PCA forces the principal components to be orthogonal (uncorrelated), which may not always be the best representation for the underlying data structure. Sometimes, important data structures exist in directions that are not orthogonal.

Not Suitable for Categorical Data:
PCA is designed for continuous, numerical data. It is not suitable for datasets with categorical variables unless those are properly encoded.

18. How do KNN and PCA complement each other?

How KNN and PCA Complement Each Other
K-Nearest Neighbors (KNN) and Principal Component Analysis (PCA) are often used together in machine learning pipelines because their strengths address each other's limitations:

Dimensionality Reduction for KNN:
KNN’s performance can degrade in high-dimensional spaces due to the curse of dimensionality, where distance metrics become less meaningful and computation becomes expensive. PCA reduces the number of features by transforming the data into a lower-dimensional space that retains most of the relevant variance, making KNN more effective and efficient.

Noise and Redundancy Removal:
PCA helps eliminate redundant and noisy features by focusing on the principal components that capture the most information. This preprocessing step improves KNN’s accuracy, as KNN is sensitive to irrelevant or redundant features.

Improved Computational Efficiency:
By reducing the number of features, PCA decreases the computational burden of KNN, which otherwise needs to calculate distances in the full feature space for every prediction.

Data Preprocessing Workflow:
A typical workflow involves applying PCA to the dataset to extract the most informative components, and then using these components as input to KNN for classification or regression. This approach is common in applications like image recognition, where raw data is high-dimensional and contains redundant information.

19. How does KNN handle missing values in a dataset?

How KNN Handles Missing Values in a Dataset
KNN (K-Nearest Neighbors) handles missing values using a technique called KNN imputation. This approach estimates and fills in missing values by leveraging the similarity between data points, as measured by a distance metric (such as Euclidean distance).

How KNN Imputation Works
Identify Neighbors:
For each data point with a missing value, KNN finds the K most similar (nearest) data points in the dataset that have observed (non-missing) values for that feature.

Estimate the Missing Value:
The missing value is then imputed using the values of the corresponding feature from these K nearest neighbors. For numerical features, the imputed value is typically the mean or weighted mean of the neighbors' values. For categorical features, it might be the most frequent category among the neighbors.

Repeat for All Missing Values:
This process is repeated for each missing value in the dataset, feature by feature.

21. Train a KNN Classifier on the Iris dataset and print model accuracy.

```python
# Train a KNN Classifier on the Iris dataset and print model accuracy

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train the KNN classifier (e.g., k=5)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = knn.predict(X_test_scaled)

# Print model accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"KNN Classifier accuracy on Iris dataset: {accuracy:.2f}")
```

22. Train a KNN Regressor on a synthetic dataset and evaluate using Mean Squared Error (MSE).

23. Train a KNN Classifier using different distance metrics (Euclidean and Manhattan) and compare accuracy.

```python
# Train a KNN Classifier using Euclidean and Manhattan distances and compare accuracy

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# KNN with Euclidean distance (default, p=2)
knn_euclidean = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn_euclidean.fit(X_train_scaled, y_train)
y_pred_euclidean = knn_euclidean.predict(X_test_scaled)
accuracy_euclidean = accuracy_score(y_test, y_pred_euclidean)

# KNN with Manhattan distance (p=1)
knn_manhattan = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=1)
knn_manhattan.fit(X_train_scaled, y_train)
y_pred_manhattan = knn_manhattan.predict(X_test_scaled)
accuracy_manhattan = accuracy_score(y_test, y_pred_manhattan)

print(f"KNN Classifier accuracy (Euclidean): {accuracy_euclidean:.2f}")
print(f"KNN Classifier accuracy (Manhattan): {accuracy_manhattan:.2f}")
```

24.  Train a KNN Classifier with different values of K and visualize decision boundaries.

```python
# Train a KNN Classifier with different values of K and visualize decision boundaries

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

# Load the iris dataset and use only the first two features for 2D visualization
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

# Define mesh step size
h = 0.02

# Define color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

# List of K values to compare
k_values = [1, 5, 15]

plt.figure(figsize=(15, 4))
for i, k in enumerate(k_values):
    # Train KNN classifier
    clf = neighbors.KNeighborsClassifier(n_neighbors=k)
    clf.fit(X, y)
    score = clf.score(X, y)

    # Create meshgrid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # Predict for each point in meshgrid
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot decision boundary and points
    plt.subplot(1, len(k_values), i + 1)
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light, shading='auto')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=40)
    plt.title(f"K = {k}\nAccuracy: {score:.2f}")
    plt.xlabel(iris.feature_names[0])
    plt.ylabel(iris.feature_names[1])
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())

plt.tight_layout()
plt.show()
```

25.  Apply Feature Scaling before training a KNN model and compare results with unscaled data.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 1. Load Data
data = load_iris()
X, y = data.data, data.target

# 2. Split Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. Feature Selection (keep top 2 features)
selector = SelectKBest(score_func=f_classif, k=2)
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected = selector.transform(X_test_scaled)

# 5. Train KNN Classifier on Selected Features
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_selected, y_train)

# 6. Evaluate
y_pred = knn.predict(X_test_selected)
accuracy = accuracy_score(y_test, y_pred)

# 7. Output Results
print("Selected feature indices:", selector.get_support(indices=True))
print("Accuracy after feature selection:", accuracy)
```

46. Train a PCA model and visualize the data reconstruction error after reducing dimensions.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error

# 1. Load and Standardize Data
data = load_iris()
X = data.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Train PCA with fewer components (e.g., 2)
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_scaled)

# 3. Reconstruct the original data from reduced form
X_reconstructed = pca.inverse_transform(X_reduced)

# 4. Compute Reconstruction Error (MSE per sample)
reconstruction_error = np.mean((X_scaled - X_reconstructed) ** 2, axis=1)

# 5. Visualization
plt.figure(figsize=(10, 5))
plt.hist(reconstruction_error, bins=20, color='skyblue', edgecolor='black')
plt.title('PCA Reconstruction Error (MSE per sample)')
plt.xlabel('Mean Squared Error')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Optional: print average error
print("Average reconstruction error (MSE):", np.mean(reconstruction_error))
```

47. Train a KNN Classifier and visualize the decision boundary.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier

# Load dataset and select first two features for 2D visualization
iris = load_iris()
X = iris.data[:, :2]
y = iris.target

# Train KNN Classifier
knn = KNeighborsClassifier(n_neighbors=15, weights='uniform')
knn.fit(X, y)

# Create meshgrid
h = 0.02  # step size
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict class for each point in the grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

# Plot decision boundary
plt.figure(figsize=(8, 6))
plt.pcolormesh(xx, yy, Z, cmap=cmap_light, shading='auto')

# Plot training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("KNN Decision Boundary (k=15, weights='uniform')")
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.show()
```

48.  Train a PCA model and analyze the effect of different numbers of components on data variance.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 1. Load and standardize the data
data = load_iris()
X = data.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Fit PCA with all components
pca = PCA()
pca.fit(X_scaled)

# 3. Extract variance explained
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

# 4. Plot explained variance and cumulative variance
plt.figure(figsize=(10, 5))

# Individual explained variance
plt.subplot(1, 2, 1)
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.7, color='steelblue')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Variance Explained by Each Component')
plt.grid(True)

# Cumulative explained variance
plt.subplot(1, 2, 2)
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', color='green')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Variance Explained')
plt.axhline(y=0.95, color='r', linestyle='--', label='95% threshold')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Optional: print exact values
print("Explained variance ratio per component:", explained_variance_ratio)
print("Cumulative variance explained:", cumulative_variance)
```
