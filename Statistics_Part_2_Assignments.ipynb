1. What is hypothesis testing in statistics?

Hypothesis testing in statistics is a formal method for making decisions or inferences about population parameters based on sample data. It allows you to test an assumption (called a hypothesis) about a population parameter.

Key Concepts
Null Hypothesis (H₀):

This is the default or initial claim.

It usually represents "no effect" or "no difference."

Example: "The average height of adult men is 70 inches."

Alternative Hypothesis (H₁ or Ha):

This is what you want to prove or the rival claim.

Example: "The average height of adult men is not 70 inches."

Test Statistic:

A number calculated from the sample data that is used to decide whether to reject H₀.

Common examples: z-score, t-score, chi-square, etc.

P-value:

The probability of observing the test statistic, or one more extreme, under the assumption that H₀ is true.

A smaller p-value indicates stronger evidence against H₀.

Significance Level (α):

The threshold for rejecting H₀, commonly set at 0.05 (5%).

If p-value ≤ α, reject H₀; otherwise, fail to reject H₀.

Conclusion:

Reject H₀: There is enough evidence to support H₁.

Fail to reject H₀: There isn’t enough evidence to support H₁, but H₀ is not proven true.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

| ---------------------- | -------------------------------- | --------------------------------------- |
| Assumption             | No effect, no difference         | There is an effect or a difference      |
| Role                   | Baseline to test against         | Claim you’re trying to find support for |
| Decision Rule          | Reject or fail to reject         | Never directly accepted                 |
| P-value interpretation | High p-value → fail to reject H₀ | Low p-value → reject H₀ (support H₁)    |


3. What is the significance level in hypothesis testing, and why is it important?

Significance Level in Hypothesis Testing
Definition:
The significance level, denoted by the Greek letter alpha (α), is the probability of rejecting the null hypothesis when it is actually true—a mistake known as a Type I error. Commonly, researchers set α at 0.05, meaning there is a 5% risk of concluding that an effect exists when there is actually none.

Why It’s Important:

Controls Error Rate: The significance level sets the threshold for how much evidence is required to reject the null hypothesis, directly controlling the probability of making a Type I error.

Guides Decision-Making: By pre-defining α, researchers have a clear rule for interpreting results: if the p-value from the data is less than α, the null hypothesis is rejected in favor of the alternative; otherwise, it is not rejected.

Standardizes Scientific Practice: Using common significance levels (such as 0.05 or 0.01) ensures consistency and comparability across studies and disciplines.

4. What does a P-value represent in hypothesis testing?

A p-value in hypothesis testing represents the probability of obtaining results at least as extreme as those observed in your data, assuming that the null hypothesis is true. In other words, it quantifies how likely it is to see the observed outcome (or something more extreme) purely by chance if there is actually no real effect or difference.

A smaller p-value indicates stronger evidence against the null hypothesis, suggesting that the observed data are unlikely under the assumption of no effect.

If the p-value is less than the chosen significance level (commonly 0.05), researchers typically reject the null hypothesis in favor of the alternative hypothesis.

Conversely, a large p-value suggests that the observed data are consistent with the null hypothesis, and there is not enough evidence to reject it.

In summary, the p-value helps determine whether the observed results are statistically significant or likely due to random chance under the null hypothesis.

5.  How do you interpret the P-value in hypothesis testing?

Interpreting the P-value in Hypothesis Testing
The p-value tells you the probability of observing data as extreme as (or more extreme than) what you actually found, assuming the null hypothesis is true.

Low p-value (typically < 0.05): Indicates that such extreme results would be very unlikely if the null hypothesis were true. This provides evidence against the null hypothesis, so you may reject it in favor of the alternative hypothesis.

High p-value (typically > 0.05): Suggests the observed data are likely under the null hypothesis. There is not enough evidence to reject the null hypothesis, so you "fail to reject" it.

6. What are Type 1 and Type 2 errors in hypothesis testing?

Type I and Type II Errors in Hypothesis Testing
Type I Error (False Positive):

Occurs when the null hypothesis is rejected even though it is actually true.

In other words, you conclude there is an effect or difference when there really isn’t.

The probability of making a Type I error is denoted by alpha (α), which is the significance level set by the researcher (commonly 0.05).

Example: Declaring a new drug effective when it is actually not.

Type II Error (False Negative):

Occurs when the null hypothesis is not rejected when it is actually false.

In other words, you fail to detect an effect or difference that truly exists.

The probability of making a Type II error is denoted by beta (β).

Example: Failing to detect that a new drug is effective when it actually is.

8. What is the Z-test, and when is it used in hypothesis testing?

What is a Z-test?
A Z-test is a statistical hypothesis test used to determine whether there is a significant difference between the means of two populations or between a sample mean and a known population mean. The test is based on the standard normal distribution (Z-distribution), and the result is expressed as a Z-score, which measures how many standard deviations a data point is from the population mean.

When is a Z-test Used?
A Z-test is appropriate under the following conditions:

The sample size is large (typically n ≥ 30), which ensures the sampling distribution of the mean is approximately normal due to the central limit theorem.

The population standard deviation (or variance) is known.

The data are approximately normally distributed.

The data points are independent of each other.

10. What is the T-distribution, and when should it be used instead of the normal distribution?

The t-distribution is a probability distribution that is symmetric and bell-shaped like the normal distribution but has heavier tails, meaning it gives more probability to extreme values. It is defined by degrees of freedom, which are related to the sample size, and its shape changes depending on these degrees of freedom—becoming more similar to the normal distribution as the sample size increases.

You should use the t-distribution instead of the normal distribution when:

The sample size is small (typically n < 30).

The population standard deviation (σ) is unknown.

You are estimating the mean of a normally distributed population using sample data.

12. What is the T-test, and how is it used in hypothesis testing?

A t-test is a statistical hypothesis test used to determine whether there is a significant difference between the means of two groups. It is commonly applied when the sample size is small and the population standard deviation is unknown. The t-test is based on the t-distribution, which accounts for the extra variability introduced by estimating the standard deviation from the sample rather than knowing it exactly.

How is the t-test used in hypothesis testing?
Formulate Hypotheses:

The null hypothesis (H₀) typically states that there is no difference between the group means.

The alternative hypothesis (H₁) suggests there is a significant difference.

Calculate the Test Statistic:

The t-test uses the means, standard deviations, and sample sizes of the groups to calculate a t-value, which measures how far apart the group means are in terms of standard error.

Compare to Critical Value:

The calculated t-value is compared to a critical value from the t-distribution (which depends on the degrees of freedom and the chosen significance level).

Interpret Results:

If the t-value exceeds the critical value, or if the p-value is less than the significance level (typically 0.05), the null hypothesis is rejected, indicating a statistically significant difference between the groups.

13. What is the relationship between Z-test and T-test in hypothesis testing?

The Z-test and T-test are both statistical methods used in hypothesis testing to determine whether there is a significant difference between sample statistics and population parameters, but they are applied under different conditions and rely on different assumptions.

Z-test: Used when the sample size is large (typically n ≥ 30) and the population variance (or standard deviation) is known. It is based on the standard normal distribution (Z-distribution). The Z-test is most precise under these conditions because the sampling distribution of the mean is approximately normal due to the Central Limit Theorem.

T-test: Used when the sample size is small (typically n < 30) and the population variance is unknown. Instead of the population variance, the sample variance is used, and the test statistic follows a t-distribution, which has heavier tails to account for the extra uncertainty in estimating the population standard deviation from the sample.

Relationship:
Both tests serve the same fundamental purpose—testing hypotheses about means—but the T-test is more robust for small samples and unknown variance, while the Z-test is appropriate for large samples with known variance. As the sample size increases, the t-distribution approaches the normal distribution, so for large samples, the results from both tests become nearly identical.

In summary, the choice between a Z-test and a T-test depends on sample size and knowledge of the population variance, but both are tools for making inferences about populations from sample data in hypothesis testing.

14. What is a confidence interval, and how is it used to interpret statistical results?

A confidence interval (CI) is a range of values, derived from sample data, that is used to estimate an unknown population parameter, such as a mean or proportion. Instead of providing just a single point estimate, a confidence interval gives a range within which the true value is likely to fall, along with a specified confidence level—commonly 95% or 99%.

For example, if a sample mean is 10 with a 95% confidence interval of 9.5 to 10.5, this means that if the sampling and interval calculation were repeated many times, about 95% of those intervals would contain the actual population mean. The confidence level reflects the long-term success rate of the method, not the probability that the true parameter is in any single calculated interval.

How is it used to interpret statistical results?

Confidence intervals quantify the uncertainty around a sample estimate, helping to assess the precision and reliability of the estimate.

In hypothesis testing, if a confidence interval for a difference between groups does not contain zero (or another value specified by the null hypothesis), it suggests a statistically significant difference.

Wider intervals indicate more uncertainty, while narrower intervals suggest more precise estimates.

Confidence intervals are commonly reported alongside p-values to provide a fuller picture of statistical results and their practical significance.

16. How is Bayes' Theorem used in statistics, and what is its significance?

Bayes' Theorem is a fundamental concept in statistics used to calculate the probability of an event based on prior knowledge of conditions related to the event. It provides a mathematical way to update or revise existing predictions or beliefs (prior probabilities) in light of new evidence (likelihood), resulting in a new, updated probability (posterior probability).

How is Bayes' Theorem used in statistics?

Bayesian Inference: One of its main applications is in Bayesian inference, where it allows statisticians to update the probability of a hypothesis as more data or evidence becomes available. For example, in medical testing, Bayes' theorem can be used to determine the probability that a patient has a disease given a positive test result, taking into account both the accuracy of the test and the prevalence of the disease in the population.

Decision-Making: It is widely used in fields such as medicine, finance, machine learning, and scientific research to make more informed decisions by incorporating both prior information and new data.

Formula: The theorem is mathematically expressed as:

$$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$$

$P(A∣B)$ is the posterior probability (probability of A given B),


$P(B∣A)$ is the likelihood (probability of B given A),


$P(A)$ is the prior probability of A,


$P(B)$ is the total probability of B.

Significance:

Updates Beliefs: Bayes' theorem is significant because it provides a rational framework for updating beliefs in the presence of new evidence, making it essential for statistical inference and decision-making.

Avoids Base-Rate Fallacy: It helps avoid common errors in interpreting probabilities, such as the base-rate fallacy, by ensuring both prior probabilities and new evidence are considered.

Wide Applications: Its versatility makes it powerful for applications ranging from evaluating the reliability of diagnostic tests to forecasting in finance and improving predictions in machine learning.

17. What is the Chi-square distribution, and when is it used?

The chi-square distribution is a continuous probability distribution that arises from the sum of the squares of independent standard normal random variables. It is characterized by its degrees of freedom, which determine its shape and spread. The distribution is defined only for positive real numbers and is widely used in inferential statistics, especially in hypothesis testing.

When is it used?

The chi-square distribution is primarily used in hypothesis tests involving categorical data. The most common applications are:

Chi-square goodness-of-fit test: Assesses how well observed categorical data fit an expected distribution.

Chi-square test for independence: Evaluates whether two categorical variables are independent in a contingency table.

It is also used in the estimation of variance and in tests involving the comparison of observed and expected frequencies.

18. What is the Chi-square goodness of fit test, and how is it applied?

The Chi-square goodness of fit test is a statistical hypothesis test used to determine whether the distribution of observed categorical data matches an expected distribution. It helps assess whether a sample comes from a population with a specific distribution or if there are significant deviations from what is theoretically expected.

How is it applied?
1. Formulate Hypotheses:

Null hypothesis (H₀): The observed frequencies fit the expected distribution (no significant difference).

Alternative hypothesis (H₁): The observed frequencies do not fit the expected distribution (significant difference).

2. Collect Data:

Gather observed counts for each category of a categorical variable.

Determine the expected counts for each category based on the hypothesized distribution.

3. Calculate the Test Statistic:

Use the formula:
$$ \chi^2 = \sum_{i=1}^{n} \frac{(O_i - E_i)^2}{E_i} $$
Where:

$O_i$ = observed frequency in category i


$E_i$ = expected frequency in category i

The sum is taken over all categories.
4. Determine Degrees of Freedom:

Degrees of freedom = (number of categories) − 1.

5. Compare to Critical Value:

Find the critical value from the Chi-square distribution table using the degrees of freedom and chosen significance level (e.g., 0.05).

If the calculated Chi-square statistic is greater than the critical value, reject the null hypothesis.

6. Interpret Results:

If you reject the null hypothesis, it suggests a significant difference between observed and expected frequencies.

If you fail to reject the null, the data fit the expected distribution well.

19. What is the F-distribution, and when is it used in hypothesis testing?

The F-distribution is a continuous, right-skewed probability distribution used primarily in hypothesis testing to compare the variances of two populations or to analyze the variability among group means in statistical models. It is defined as the ratio of two independent chi-square distributions, each divided by their respective degrees of freedom. The shape of the F-distribution depends on two parameters: the degrees of freedom for the numerator and the denominator.

When is it used in hypothesis testing?

The F-distribution is most commonly used in the analysis of variance (ANOVA), where it tests whether the means of two or more groups are significantly different by comparing the variance between groups to the variance within groups.

It is also used to test the equality of two variances, such as in the F-test for comparing the variability of two samples.

More broadly, any situation where the ratio of two sample variances is of interest may involve the F-distribution.

20. What is an ANOVA test, and what are its assumptions?

An ANOVA (Analysis of Variance) test is a statistical method used to determine whether there are statistically significant differences between the means of three or more groups. Unlike multiple t-tests, which increase the risk of Type I error, ANOVA allows for simultaneous comparison of group means by analyzing the variance both between and within groups. The test produces an F-statistic, which quantifies the ratio of variance between groups to variance within groups; a higher F-value suggests greater differences among group means.

Types of ANOVA
One-way ANOVA: Compares means across groups based on one independent variable.

Two-way ANOVA: Examines the effect of two independent variables and their interaction on a dependent variable.

Assumptions of ANOVA
Independence: Observations must be independent of each other.

Normality: The data within each group should be approximately normally distributed.

Homogeneity of Variances: The variances among the groups should be roughly equal.

21. What are the different types of ANOVA tests?

There are several types of ANOVA (Analysis of Variance) tests, each suited to different experimental designs and research questions:

Main Types of ANOVA Tests
One-Way ANOVA:
Compares the means of three or more independent groups based on a single factor (independent variable). For example, it can test whether different teaching methods produce different average test scores among students.

Two-Way ANOVA:
Examines the effect of two independent variables (factors) on a single continuous outcome, and can also assess whether there is an interaction between the two factors. For example, it might analyze the combined effects of teaching method and gender on student performance.

Factorial ANOVA:
An extension of two-way ANOVA that involves more than two factors, allowing for the analysis of complex experimental designs with multiple independent variables and their interactions.

Repeated Measures ANOVA:
Used when the same subjects are measured multiple times under different conditions or over time (within-subjects design). For example, measuring blood pressure in the same group of patients at several time points.

Mixed-Model ANOVA (Within-Between ANOVA):
Combines features of both between-subjects and within-subjects designs, analyzing data with both independent groups and repeated measurements on the same subjects.

Welch’s ANOVA:
Used when the assumption of equal variances across groups is violated.

Ranked ANOVA:
Applied when data are ordinal or when assumptions of standard ANOVA are not met, using rank-transformed data.

22. What is the F-test, and how does it relate to hypothesis testing?

The F-test is a statistical hypothesis test used to compare the variances of two populations or samples to determine if they are significantly different. It calculates the F-statistic, which is the ratio of the variances (usually the larger variance divided by the smaller variance), and this statistic follows an F-distribution under the null hypothesis.

In hypothesis testing, the F-test is commonly used in several contexts:

Testing equality of variances: To check if two populations have the same variance, which is important before performing other tests like the t-test.

Analysis of Variance (ANOVA): The F-test is the basis for ANOVA, where it assesses whether the variability between group means is significantly greater than the variability within groups, indicating that at least one group mean is different from the others.

Regression analysis: The F-test is used to determine if a regression model as a whole is statistically significant, i.e., whether the independent variables explain a significant portion of the variance in the dependent variable.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
import numpy as np
from statsmodels.stats.weightstats import ztest

# Example data: sample of test scores
sample_data = [78, 82, 85, 90, 76, 88, 79, 81, 77, 83, 80, 84, 86, 87, 82, 81, 79, 80, 85, 83]
population_mean = 75     # Known population mean
population_std = 10      # Known population standard deviation (if unknown, use t-test instead)
alpha = 0.05             # Significance level

# Perform the Z-test
z_stat, p_value = ztest(sample_data, value=population_mean, alternative='two-sided', ddof=0)

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")
```

3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
import numpy as np
from scipy.stats import norm

# --- Step 1: Simulate or provide sample data ---
np.random.seed(0)
sample = np.random.normal(loc=52, scale=5, size=40)  # Example: mean=52, std=5, n=40

# --- Step 2: Specify population parameters ---
population_mean = 50      # Known population mean
population_std = 5        # Known population standard deviation
n = len(sample)

# --- Step 3: Calculate sample mean ---
sample_mean = np.mean(sample)

# --- Step 4: Calculate Z-statistic ---
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# --- Step 5: Calculate two-tailed P-value ---
p_value = 2 * (1 - norm.cdf(abs(z)))

# --- Step 6: Output results ---
print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-statistic: {z:.3f}")
print(f"P-value: {p_value:.4f}")

# --- Step 7: Interpret the result ---
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate sample data
np.random.seed(1)
sample = np.random.normal(loc=52, scale=5, size=40)  # mean=52, std=5, n=40

# Known population parameters
population_mean = 50
population_std = 5
n = len(sample)

# Calculate sample mean and Z-statistic
sample_mean = np.mean(sample)
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Two-tailed test: calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Critical Z-values for alpha = 0.05 (two-tailed)
alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)  # ≈ ±1.96

# Print results
print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Z-values: ±{z_critical:.2f}")

# Interpretation
if abs(z_stat) > z_critical:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")

# Visualization
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution')

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.5, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.5, label='Rejection Region (Right)')

# Plot observed Z-statistic
plt.axvline(z_stat, color='blue', linestyle='--', linewidth=2, label=f'Observed Z = {z_stat:.2f}')

# Add critical values
plt.axvline(-z_critical, color='black', linestyle=':', linewidth=1, label=f'Critical Z = -{z_critical:.2f}')
plt.axvline(z_critical, color='black', linestyle=':', linewidth=1, label=f'Critical Z = +{z_critical:.2f}')

plt.title('Two-Tailed Z-test: Decision Regions')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(
    mu0=0,         # Null hypothesis mean
    mu1=1,         # Alternative hypothesis mean
    sigma=1,       # Known population standard deviation
    n=30,          # Sample size
    alpha=0.05,    # Significance level
    two_tailed=True
):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical values for rejection region(s)
    if two_tailed:
        z_crit = norm.ppf(1 - alpha/2)
        crit_right = mu0 + z_crit * se
        crit_left = mu0 - z_crit * se
    else:
        z_crit = norm.ppf(1 - alpha)
        crit_right = mu0 + z_crit * se
        crit_left = None

    # X range for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    null_dist = norm.pdf(x, mu0, se)
    alt_dist = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist, label="Null Hypothesis ($H_0$)", color="blue")
    plt.plot(x, alt_dist, label="Alternative Hypothesis ($H_1$)", color="green")

    # Shade Type I error region(s) (alpha): Reject H0 when H0 is true
    if two_tailed:
        plt.fill_between(x, 0, null_dist, where=(x <= crit_left) | (x >= crit_right), color='red', alpha=0.3, label="Type I Error (α)")
    else:
        plt.fill_between(x, 0, null_dist, where=(x >= crit_right), color='red', alpha=0.3, label="Type I Error (α)")

    # Shade Type II error region (beta): Fail to reject H0 when H1 is true
    if two_tailed:
        plt.fill_between(x, 0, alt_dist, where=(x > crit_left) & (x < crit_right), color='orange', alpha=0.3, label="Type II Error (β)")
    else:
        plt.fill_between(x, 0, alt_dist, where=(x < crit_right), color='orange', alpha=0.3, label="Type II Error (β)")

    # Draw critical value lines
    if two_tailed:
        plt.axvline(crit_left, color='black', linestyle='--', label=f'Critical Value Left ({crit_left:.2f})')
    plt.axvline(crit_right, color='black', linestyle='--', label=f'Critical Value Right ({crit_right:.2f})')

    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Calculate and print Type I and Type II error probabilities
    if two_tailed:
        type2_prob = norm.cdf(crit_right, loc=mu1, scale=se) - norm.cdf(crit_left, loc=mu1, scale=se)
    else:
        type2_prob = norm.cdf(crit_right, loc=mu1, scale=se)
    print(f"Type I Error (α): {alpha:.3f}")
    print(f"Type II Error (β): {type2_prob:.3f}")

# Example usage:
visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05, two_tailed=True)
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        sample (array-like): The sample data.
        confidence (float): The confidence level (default is 0.95 for 95%).

    Returns:
        (lower_bound, upper_bound): Tuple containing the confidence interval.
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # Sample standard error
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)   # Critical t-value
    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return lower_bound, upper_bound

# Example usage
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=25)
ci_low, ci_high = confidence_interval(data, confidence=0.95)
print(f"95% Confidence Interval for the mean: ({ci_low:.2f}, {ci_high:.2f})")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for a sample mean at a given confidence level.

    Parameters:
        sample (array-like): The sample data.
        confidence (float): The confidence level (default is 0.95 for 95%).

    Returns:
        float: The margin of error.
    """
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # Standard error of the mean
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)   # Critical t-value
    margin = t_crit * std_err
    return margin

# Example usage
np.random.seed(1)
data = np.random.normal(loc=75, scale=8, size=30)
moe = margin_of_error(data, confidence=0.95)
print(f"Margin of Error (95% confidence): {moe:.2f}")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior, sensitivity, specificity):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
        prior (float): Prior probability of disease (prevalence).
        sensitivity (float): Probability of a positive test given disease (true positive rate).
        specificity (float): Probability of a negative test given no disease (true negative rate).

    Returns:
        posterior (float): Probability of disease given a positive test.
    """
    # Probability of a positive test given disease is present
    p_pos_given_disease = sensitivity
    # Probability of a positive test given disease is absent (false positive rate)
    p_pos_given_no_disease = 1 - specificity

    # Marginal probability of a positive test
    p_pos = (p_pos_given_disease * prior) + (p_pos_given_no_disease * (1 - prior))

    # Posterior probability using Bayes' Theorem
    posterior = (p_pos_given_disease * prior) / p_pos
    return posterior

# Example: Medical test scenario
prior = 0.01          # 1% prevalence
sensitivity = 0.95    # 95% true positive rate
specificity = 0.98    # 98% true negative rate

posterior = bayes_theorem(prior, sensitivity, specificity)
print(f"Probability of disease given a positive test: {posterior:.3%}")
```

16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy.stats import f

# Example random samples
np.random.seed(0)
sample1 = np.random.normal(loc=10, scale=4, size=30)  # mean=10, std=4, n=30
sample2 = np.random.normal(loc=10, scale=2, size=30)  # mean=10, std=2, n=30

# Calculate sample variances (use ddof=1 for sample variance)
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# F-test statistic (larger variance / smaller variance)
if var1 > var2:
    f_stat = var1 / var2
    dfn = len(sample1) - 1  # numerator degrees of freedom
    dfd = len(sample2) - 1  # denominator degrees of freedom
else:
    f_stat = var2 / var1
    dfn = len(sample2) - 1
    dfd = len(sample1) - 1

# Calculate p-value (right-tailed)
p_value = 1 - f.cdf(f_stat, dfn, dfd)

print(f"Sample 1 variance: {var1:.3f}")
print(f"Sample 2 variance: {var2:.3f}")
print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
from scipy.stats import shapiro, levene
import matplotlib.pyplot as plt

def check_anova_assumptions(*groups, plot=True):
    """
    Check ANOVA assumptions: normality, homogeneity of variances, and independence.
    Parameters:
        *groups: arrays of sample data for each group
        plot: if True, creates QQ-plots for normality check
    Returns:
        None (prints results)
    """
    print("Checking ANOVA assumptions...\n")

    # 1. Normality (Shapiro-Wilk test for each group)
    print("Normality (Shapiro-Wilk test):")
    for i, group in enumerate(groups, 1):
        stat, p = shapiro(group)
        print(f"  Group {i}: W={stat:.3f}, p-value={p:.4f} {'(Normal)' if p>0.05 else '(Not normal)'}")
        if plot:
            plt.figure()
            from scipy.stats import probplot
            probplot(group, dist="norm", plot=plt)
            plt.title(f"QQ-Plot Group {i}")
    if plot:
        plt.show()

    # 2. Homogeneity of variances (Levene’s test)
    stat, p = levene(*groups)
    print(f"\nHomogeneity of variances (Levene's test): W={stat:.3f}, p-value={p:.4f} {'(Equal variances)' if p>0.05 else '(Variances differ)'}")

    # 3. Independence
    print("\nIndependence:")
    print("  Independence is typically ensured by study design (e.g., random sampling).")
    print("  If data are not paired/repeated measures and collected randomly, this assumption is likely met.")
    print("  For time series or sequential data, plot residuals or use the Durbin-Watson test if needed.")

# Example usage:
np.random.seed(0)
group1 = np.random.normal(50, 5, 30)
group2 = np.random.normal(52, 5, 30)
group3 = np.random.normal(54, 5, 30)

check_anova_assumptions(group1, group2, group3, plot=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Example data: Two random samples
np.random.seed(0)
data1 = np.random.normal(loc=20, scale=5, size=30)
data2 = np.random.normal(loc=22, scale=2, size=30)

# Calculate sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Ensure the larger variance is in the numerator
if var1 >= var2:
    f_stat = var1 / var2
    dfn = len(data1) - 1  # degrees of freedom numerator
    dfd = len(data2) - 1  # degrees of freedom denominator
else:
    f_stat = var2 / var1
    dfn = len(data2) - 1
    dfd = len(data1) - 1

# Calculate the p-value (two-tailed)
p_value = 2 * min(1 - f.cdf(f_stat, dfn, dfd), f.cdf(f_stat, dfn, dfd))

print(f"Sample 1 variance: {var1:.3f}")
print(f"Sample 2 variance: {var2:.3f}")
print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")

# Visualization
plt.figure(figsize=(8,5))
plt.hist(data1, bins=15, alpha=0.6, label='Sample 1', color='skyblue', density=True)
plt.hist(data2, bins=15, alpha=0.6, label='Sample 2', color='salmon', density=True)
plt.title('Distributions of the Two Samples')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# --- 1. Simulate Observed Data ---
# Let's simulate rolling a 6-sided die 600 times.
# If the die were perfectly fair, we'd expect each side to appear 100 times.
# We'll introduce some variation to simulate real-world observations.

np.random.seed(42) # for reproducibility

# Simulate observed frequencies for a 6-sided die
# (e.g., counts for rolling 1, 2, 3, 4, 5, 6)
observed_frequencies = np.array([95, 110, 85, 105, 100, 105])
total_observations = np.sum(observed_frequencies)
print(f"Observed Frequencies: {observed_frequencies}")
print(f"Total Observations: {total_observations}\n")

# --- 2. Define Expected Proportions/Frequencies ---
# Null Hypothesis (H0): The die is fair.
# This means each side should have an equal probability (1/6).

num_categories = len(observed_frequencies)
expected_proportions = np.array([1/num_categories] * num_categories) # [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

# Calculate expected frequencies based on total observations
expected_frequencies = expected_proportions * total_observations
print(f"Expected Proportions (under H0): {expected_proportions}")
print(f"Expected Frequencies (under H0): {expected_frequencies}\n")

# --- 3. Perform Chi-square Test for Goodness of Fit ---
# We use scipy.stats.chisquare(f_obs, f_exp=None)
# f_obs: array of observed frequencies
# f_exp: array of expected frequencies (if None, assumes uniform distribution)

chi2_statistic, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

print(f"Chi-square Statistic: {chi2_statistic:.4f}")
print(f"P-value: {p_value:.4f}\n")

# --- 4. Analyze the Results ---

# Define a significance level (alpha)
alpha = 0.05

print("--- Analysis of Results ---")
print(f"Null Hypothesis (H0): The observed frequencies fit the expected distribution (the die is fair).")
print(f"Alternative Hypothesis (H1): The observed frequencies do not fit the expected distribution (the die is not fair).\n")
print(f"Chosen Significance Level (alpha): {alpha}")

if p_value < alpha:
    print(f"P-value ({p_value:.4f}) is less than alpha ({alpha}).")
    print("Conclusion: We reject the null hypothesis.")
    print("There is sufficient evidence to suggest that the observed frequencies do not significantly fit the expected uniform distribution (the die is likely not fair).")
else:
    print(f"P-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}).")
    print("Conclusion: We fail to reject the null hypothesis.")
    print("There is not sufficient evidence to suggest that the observed frequencies significantly differ from the expected uniform distribution (the die appears to be fair based on this test).")

# --- Optional: Visualize the Frequencies ---
categories = [str(i) for i in range(1, num_categories + 1)]
x = np.arange(num_categories)
width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))
rects1 = ax.bar(x - width/2, observed_frequencies, width, label='Observed')
rects2 = ax.bar(x + width/2, expected_frequencies, width, label='Expected')

ax.set_xlabel('Die Face')
ax.set_ylabel('Frequency')
ax.set_title('Observed vs. Expected Frequencies of Die Rolls')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add frequency values on top of bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{int(height)}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()
plt.show()
```
