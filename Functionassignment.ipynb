7. What are the advantages of using generators over regular functions?

Ans: Generators have several advantages over regular functions, including:
1) Memory efficiency: Generators only evaluate code when needed, so they don't need to store all values in memory at once.
2) Lazy loading: Generators can load data when it's needed, which can improve performance and reduce bandwidth usage.
3) Flexibility: Generators offer a more flexible approach to computation than regular functions.
4) Infinite sequences: Generators can create infinite sequences by allowing pausing and resuming.
5) Readability: Generators can make code easier to read and understand by breaking up the iteration process into smaller chunks.
6) Speed: Generators can be faster than regular loops, especially when working with large data sets.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans: The Map function, from a wide perspective, is about transformation. It applies a given function to each item of an array, creating a new array with the results.
The Filter function aims at selection. It constructs a new array that includes elements of the original array that meet a specified condition.
The Reduce function, as its name suggests, focuses on reduction. It processes an array to combine its values and reduce them down to a single value.

```python
# map() example
l1 = [100, 200,300, 400]
l2 = [1, 2, 3, 4]
list(map(lambda x, y : x+y, l1, l2))
```

```python
#reduce() example
from functools import reduce
l = [2, 1, 3, 4, 4, 5, 6]
reduce(lambda x, y:x+y, l)
```

```python
# filter() example
l = [2, 1, 3, 4, 4, 5, 6]
list(filter(lambda x: x%2 == 0, l))
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def even_sum(numbers):
  sum=0
  for n in numbers:
    if n%2==0:
        sum=sum+n
  return(sum)

even_sum([1,2,3,4,5,6])
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, limit):
        self.limit = limit
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.a > self.limit:
            raise StopIteration
        current = self.a
        self.a, self.b = self.b, self.a + self.b
        return current
fib_iterator = FibonacciIterator(50)
print(next(fib_iterator))
print(next(fib_iterator))
print(next(fib_iterator))
print(next(fib_iterator))
print(next(fib_iterator))
print(next(fib_iterator))
print(next(fib_iterator))
print(next(fib_iterator))
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(n):
    for i in range(n):
        yield 2**i
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    result = ''.join([char for char in input_string if char not in vowels])
    return result
#Example usage
input_string = "Hello, World!"
output_string = remove_vowels(input_string)
print("Original String:", input_string)
print("String without vowels:", output_string)



```
