5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set the desired number of features and sample size
num_features = 5
sample_size = 1000

# Create a covariance matrix with varying degrees of correlation
cov_matrix = np.array([[1, 0.8, 0.5, 0.2, 0.1],
                       [0.8, 1, 0.7, 0.4, 0.3],
                       [0.5, 0.7, 1, 0.6, 0.5],
                       [0.2, 0.4, 0.6, 1, 0.7],
                       [0.1, 0.3, 0.5, 0.7, 1]])

# Generate the synthetic dataset
data = np.random.multivariate_normal(mean=np.zeros(num_features), cov=cov_matrix, size=sample_size)

# Convert the data to a DataFrame for easier visualization
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5'])

# Compute the correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

# Create a pivot table to aggregate sales data by month and day
pivot_df = df.pivot_table(index='Month', columns='Day', values='Sales')

# Create a heatmap using Plotly Express
fig = px.imshow(pivot_df,
                x=pivot_df.columns,
                y=pivot_df.index,
                color_continuous_scale='Viridis',
                title='Sales Variation by Month and Day')

# Customize the heatmap
fig.update_layout(
    xaxis_title='Day',
    yaxis_title='Month',
    coloraxis_colorbar_title='Sales',
    font_family='Arial',
    font_size=12
)

# Show the heatmap
fig.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Given data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(x=df['X'], y=df['Y'], z=df['Z'], colorscale='viridis', opacity=0.8)])

# Customize the plot
fig.update_layout(
    title='3D Surface Plot of sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

# Show the plot
fig.show()
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()


import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool

# Generate random data
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)
sizes = np.random.randint(10, 30, num_points)
colors = np.random.choice(['red', 'blue', 'green'], num_points)

# Create a pandas DataFrame
data = {'x': x, 'y': y, 'sizes': sizes, 'colors': colors}
df = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create the figure
p = figure(title="Scatter Plot with Random Data")

# Add a scatter plot
scatter = p.scatter(x='x', y='y', size='sizes', color='colors', source=source)

# Add a tooltip
tooltip = HoverTool(tooltips=[('x', '@x'), ('y', '@y'), ('size', '@sizes'), ('color', '@colors')])
p.add_tools(tooltip)

# Show the plot
show(p)
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Step 1: Generate Data
data_hist = np.random.randn(1000)

# Step 2: Calculate Histogram
hist, edges = np.histogram(data_hist, bins=30)

# Step 3: Create a Bokeh Plot
output_notebook()  # Use this if you're running in Jupyter Notebook

p = figure(title='Histogram of Normally Distributed Data',
           x_axis_label='Value',
           y_axis_label='Frequency')

# Step 4: Add Quad Glyphs for Histogram Bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white', alpha=0.5)

# Step 5: Show the Plot
show(p)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
from bokeh.io import output_notebook

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data for Bokeh
df = pd.DataFrame(data_heatmap, columns=x, index=y)
df = df.stack().rename("value").reset_index()
df.columns = ['x', 'y', 'value']

# Create a color mapper
mapper = LinearColorMapper(palette="Viridis256", low=df.value.min(), high=df.value.max())

# Create the figure
p = figure(title="Heatmap Example", x_axis_label='X', y_axis_label='Y', tools="hover", tooltips=[("Value", "@value")])
p.rect(x="x", y="y", width=1, height=1, source=df, line_color=None, fill_color=transform('value', mapper))

# Add color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
output_notebook()
show(p)
```
